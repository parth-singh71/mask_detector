```python
import glob, os
import random
```

```python
directory = os.getcwd()
dataset_path = directory+'/data/dataset/images/' # path to image data (not labels)

train_percent = 80
# test_percent = 10
validation_percent = 20
```

```python
dataset_path
```

```python
files = os.listdir(dataset_path)
get_abs_path = lambda path : dataset_path+path
files = list(map(get_abs_path, files))
```

```python
files
```

```python
length = len(files)
length
```

```python
def get_remaining_files(data, used_data):
    res = []
    for file in data:
        if file not in used_data:
            res.append(file)
    return res
```

```python
def save_lists_to_files(train_set, validation_set):
    with open("data/train.txt", "w+") as f:
        f.write("\n".join(train_set))

    with open("data/val.txt", "w+") as f:
        f.write("\n".join(validation_set))
```

```python
def split_dataset(data, train_percent, validation_percent):
    k1 = round(length * (train_percent / 100))
    train_set = random.sample(data, k1)
    validation_set = get_remaining_files(data, train_set)
    save_lists_to_files(train_set, validation_set)
    return train_set, validation_set
```

```python
split_dataset(files, train_percent, validation_percent)
```
